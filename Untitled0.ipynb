```python
print(" Kurulum (1 kere) ...")
!pip -q install -U google-generativeai pymupdf chromadb rank-bm25 sentence-transformers transformers accelerate torch pandas tqdm

import os, re, json, time
import fitz  # PyMuPDF
import torch
import pandas as pd
import google.generativeai as genai
import chromadb
from chromadb.config import Settings
from dataclasses import dataclass
from tqdm import tqdm
from rank_bm25 import BM25Okapi
from sentence_transformers import SentenceTransformer
from transformers import AutoTokenizer, AutoModelForSequenceClassification

PDF_PATH = "/content/7.5.39586.pdf"
LOG_DIR = "/content/logs"
os.makedirs(LOG_DIR, exist_ok=True)

SLEEP_SEC = 13
MAX_RETRIES = 6
BACKOFF_BASE = 2.0
RETRY_IN_RE = re.compile(r"Please retry in\s+([0-9.]+)s", re.IGNORECASE)

try:
    from google.colab import userdata
    GEMINI_API_KEY = userdata.get("GEMINI_API_KEY")
    if not GEMINI_API_KEY:
        raise ValueError("Secrets'ta GEMINI_API_KEY yok.")
    print(" API key: Colab Secrets'tan alÄ±ndÄ±.")
except Exception:
    GEMINI_API_KEY = input("GEMINI_API_KEY gir: ").strip()

genai.configure(api_key=GEMINI_API_KEY)

def _normalize_model_name(name: str) -> str:
    return (name or "").replace("models/", "").strip()

def pick_working_model(preferred=None):
    preferred = preferred or [
        "gemini-2.5-flash",
        "gemini-2.0-flash",
        "gemini-2.0-flash-lite",
        "gemini-1.5-flash-latest",
        "gemini-1.5-flash",
        "gemini-1.5-pro-latest",
        "gemini-1.5-pro",
    ]
    models = list(genai.list_models())
    supported = []
    for m in models:
        methods = getattr(m, "supported_generation_methods", None) or getattr(m, "supportedGenerationMethods", None) or []
        if "generateContent" in methods:
            supported.append(_normalize_model_name(m.name))
    if not supported:
        raise RuntimeError("HiÃ§bir model generateContent desteklemiyor. API key/eriÅŸim yanlÄ±ÅŸ olabilir.")
    for cand in preferred:
        if cand in supported:
            return cand, supported
    return supported[0], supported

MODEL_NAME, AVAILABLE_MODELS = pick_working_model()
print(" SeÃ§ilen model:", MODEL_NAME)

def robust_generate(model, prompt: str) -> str:
    for attempt in range(1, MAX_RETRIES + 1):
        try:
            return model.generate_content(prompt).text
        except Exception as e:
            msg = str(e)
            m = RETRY_IN_RE.search(msg)
            retry_in = float(m.group(1)) if m else None

            is_429 = ("429" in msg) or ("quota" in msg.lower()) or ("ResourceExhausted" in msg)
            is_conn = ("Connection reset" in msg) or ("RemoteDisconnected" in msg) or ("Connection aborted" in msg)

            if is_429 or is_conn:
                wait = retry_in if retry_in is not None else (BACKOFF_BASE ** attempt)
                wait = max(wait, SLEEP_SEC)
                print(f" GeÃ§ici hata (attempt {attempt}/{MAX_RETRIES}). Bekleme: {wait:.1f}s")
                time.sleep(wait)
                continue
            raise
    return "Hata: Quota/baÄŸlantÄ± sorunu devam ediyor (Ã§ok denendi)."

try:
    _ping_model = genai.GenerativeModel(MODEL_NAME)
    _pong = robust_generate(_ping_model, "ping").strip()[:50]
    print(" Model ping OK:", _pong)
except Exception as e:
    raise RuntimeError(f"Model Ã§aÄŸrÄ±lamadÄ±: {MODEL_NAME}\nHata: {e}")

print(f"\n PDF kontrol: {PDF_PATH}")
if not os.path.exists(PDF_PATH):
    raise FileNotFoundError(f"PDF bulunamadÄ±: {PDF_PATH}\nDosyayÄ± Colab'a yÃ¼kleyip /content iÃ§ine koy.")

@dataclass
class Chunk:
    chunk_id: str
    text: str
    source: str
    page_start: int
    page_end: int
    madde_no: str | None

MADDE_RE = re.compile(r"(?mi)^\s*MADDE\s+(\d+)\s*(?:[-â€“:.\)]\s*)?", flags=re.IGNORECASE | re.MULTILINE)
PAGE_MARK_RE = re.compile(r"\[\[\[PAGE:(\d+)\]\]\]")

def extract_and_chunk(pdf_path: str, max_chars=2000, overlap=200):
    doc = fitz.open(pdf_path)
    pages = []
    for i in range(len(doc)):
        txt = doc[i].get_text("text") or ""
        txt = re.sub(r"[ \t]+", " ", txt)
        txt = re.sub(r"\n{3,}", "\n\n", txt).strip()
        pages.append((i + 1, txt))

    full = []
    for pno, txt in pages:
        full.append(f"\n\n[[[PAGE:{pno}]]]\n\n{txt}")
    full_text = "".join(full)

    starts = [(m.start(), m.group(1)) for m in MADDE_RE.finditer(full_text)]
    if not starts:
        starts = [(0, None)]
    starts.append((len(full_text), None))

    out = []
    k = 0
    source_name = os.path.basename(pdf_path)

    for i in range(len(starts) - 1):
        s, madde = starts[i]
        e, _ = starts[i + 1]
        block = full_text[s:e].strip()
        if not block:
            continue

        pages_in_block = [int(x) for x in PAGE_MARK_RE.findall(block)]
        p_start = min(pages_in_block) if pages_in_block else 1
        p_end = max(pages_in_block) if pages_in_block else p_start
        clean = PAGE_MARK_RE.sub("", block).strip()

        if len(clean) <= max_chars:
            out.append(Chunk(f"m{madde or 'X'}_{k}", clean, source_name, p_start, p_end, madde))
            k += 1
        else:
            pos = 0
            while pos < len(clean):
                end = min(len(clean), pos + max_chars)
                sub = clean[pos:end]
                out.append(Chunk(f"m{madde or 'X'}_{k}", sub, source_name, p_start, p_end, madde))
                k += 1
                if end == len(clean):
                    break
                pos = max(0, end - overlap)
    return out

print(" Chunking...")
chunks = extract_and_chunk(PDF_PATH)
print(f" Chunk sayÄ±sÄ±: {len(chunks)}")

print("\n Modeller yÃ¼kleniyor (Embedding + Reranker)...")

embedder = SentenceTransformer("intfloat/multilingual-e5-base")

client_db = chromadb.Client(Settings(anonymized_telemetry=False))
try:
    client_db.delete_collection("kb_docs")
except Exception:
    pass
col = client_db.get_or_create_collection("kb_docs")

print(" Chroma index yazÄ±lÄ±yor...")
BATCH = 64
for i in range(0, len(chunks), BATCH):
    batch = chunks[i:i+BATCH]
    embs = embedder.encode([f"passage: {c.text}" for c in batch], normalize_embeddings=True).tolist()
    col.add(
        ids=[c.chunk_id for c in batch],
        documents=[c.text for c in batch],
        metadatas=[{"m_no": c.madde_no, "p_start": c.page_start, "p_end": c.page_end} for c in batch],
        embeddings=embs
    )

def tok_tr(s: str):
    return [t for t in re.sub(r"[^a-zÃ§ÄŸÄ±Ã¶ÅŸÃ¼0-9\s]", " ", (s or "").lower()).split() if len(t) > 1]

bm25 = BM25Okapi([tok_tr(c.text) for c in chunks])

rerank_tokenizer = AutoTokenizer.from_pretrained("BAAI/bge-reranker-v2-m3")
rerank_model = AutoModelForSequenceClassification.from_pretrained("BAAI/bge-reranker-v2-m3")
device = "cuda" if torch.cuda.is_available() else "cpu"
rerank_model.to(device).eval()
print(f" Reranker OK | device={device}")

@torch.no_grad()
def rerank_scores(query, candidates):
    if not candidates:
        return []
    scores_all = []
    for i in range(0, len(candidates), 8):
        batch = candidates[i:i+8]
        pairs = [[query, c] for c in batch]
        inputs = rerank_tokenizer(pairs, padding=True, truncation=True, max_length=512, return_tensors="pt").to(device)
        logits = rerank_model(**inputs).logits.view(-1).float().cpu().tolist()
        scores_all.extend(logits)
    return scores_all

def retrieve_ranked(query: str, dense_k=15, bm25_k=15, final_k=3):
    dense_res = col.query(
        query_embeddings=embedder.encode([f"query: {query}"], normalize_embeddings=True).tolist(),
        n_results=dense_k
    )
    dense_docs = dense_res["documents"][0] if dense_res.get("documents") else []
    dense_meta = dense_res["metadatas"][0] if dense_res.get("metadatas") else []

    bm25_scores = bm25.get_scores(tok_tr(query))
    bm25_idx = sorted(range(len(bm25_scores)), key=lambda i: bm25_scores[i], reverse=True)[:bm25_k]
    bm25_docs = [chunks[i].text for i in bm25_idx]
    bm25_meta = [{"m_no": chunks[i].madde_no, "p_start": chunks[i].page_start, "p_end": chunks[i].page_end} for i in bm25_idx]

    candidates = {}
    for d, m in zip(dense_docs, dense_meta):
        candidates[d] = m
    for d, m in zip(bm25_docs, bm25_meta):
        candidates[d] = m

    unique_docs = list(candidates.keys())
    unique_metas = [candidates[d] for d in unique_docs]
    if not unique_docs:
        return []

    scores = rerank_scores(query, unique_docs[:20])
    ranked = sorted(zip(unique_docs, unique_metas, scores), key=lambda x: x[2], reverse=True)[:final_k]
    return ranked

def search_tool(query: str) -> str:
    results = retrieve_ranked(query)
    if not results:
        return "SonuÃ§ bulunamadÄ±."
    ctx = []
    for i, (txt, meta, scr) in enumerate(results, 1):
        ctx.append(f"[KANIT {i}] (Madde: {meta.get('m_no')}, Sayfa: {meta.get('p_start')})\n{txt[:900]}")
    return "\n\n".join(ctx)

SYSTEM_RULES = """
Kurallar:
- SADECE verilen KANIT'a dayan.
- KANIT iÃ§inde yoksa: "Bilgi bulunamadÄ±." de.
- Uydurma yok.
- KÄ±sa ve net cevap ver.
"""

QA_MODEL = genai.GenerativeModel(MODEL_NAME)

def answer_with_rag(question: str) -> str:
    evidence = search_tool(question)
    prompt = f"""{SYSTEM_RULES}

KANIT:
{evidence}

SORU:
{question}

CEVAP:
"""
    return robust_generate(QA_MODEL, prompt).strip()

def normalize(s: str) -> str:
    s = (s or "").lower()
    s = s.replace("â‚º", "tl")
    s = re.sub(r"\s+", " ", s).strip()
    return s

def score_answer(qtype: str, model_ans: str, gt: str):
    a = normalize(model_ans)
    g = normalize(gt)

    if "edge" in qtype.lower() or "refusal" in qtype.lower():
        refusal_ok = any(k in a for k in ["bilgi bulunamadÄ±", "kapsam dÄ±ÅŸÄ±", "bilmiyorum", "yok", "bulunamadÄ±", "refusal"])
        return (100, "DoÄŸru reddetti.") if refusal_ok else (0, "Reddetmesi gerekirken cevap verdi/uydurdu.")

    if g in a:
        return 100, "GT aynen geÃ§ti."

    nums_g = re.findall(r"\d[\d\.\,]*", g)
    nums_a = re.findall(r"\d[\d\.\,]*", a)
    if nums_g:
        clean = lambda x: re.sub(r"[^\d]", "", x)
        cg = [clean(x) for x in nums_g if clean(x)]
        ca = [clean(x) for x in nums_a if clean(x)]
        if any(x in ca for x in cg):
            return 100, "SayÄ±sal deÄŸer eÅŸleÅŸti."

    if len(g) <= 20 and any(tok in a for tok in g.split()):
        return 100, "KÄ±sa GT anahtar kelime ile yakalandÄ±."

    return 0, "EÅŸleÅŸme yok."

benchmark_data = [
  {"id":"N01","type":"Factual","q":"TMFB belgesinin geÃ§erlilik sÃ¼resi kaÃ§ yÄ±ldÄ±r?","gt":"TODO"},
  {"id":"N02","type":"Factual","q":"TMFB baÅŸvurusu hangi kurum/otorite Ã¼zerinden yapÄ±lÄ±r?","gt":"TODO"},
  {"id":"N03","type":"Factual","q":"TMFB belgesinin yenilenmesi iÃ§in hangi koÅŸullar aranÄ±r?","gt":"TODO"},
  {"id":"N04","type":"Factual","q":"TMGD hizmeti alma zorunluluÄŸu hangi durumlarda doÄŸar?","gt":"TODO"},
  {"id":"N05","type":"Factual","q":"TMGD hizmeti alma zorunluluÄŸu olmayan iÅŸletmelerin eÄŸitimi kaÃ§ yÄ±lda bir yenilenir?","gt":"TODO"},
  {"id":"N06","type":"Factual","q":"Tehlikeli madde taÅŸÄ±macÄ±lÄ±ÄŸÄ±nda 'gÃ¶nderen'in temel sorumluluklarÄ± nelerdir?","gt":"TODO"},
  {"id":"N07","type":"Factual","q":"Tehlikeli madde taÅŸÄ±macÄ±lÄ±ÄŸÄ±nda 'taÅŸÄ±yan'Ä±n temel sorumluluklarÄ± nelerdir?","gt":"TODO"},
  {"id":"N08","type":"Factual","q":"Tehlikeli madde taÅŸÄ±macÄ±lÄ±ÄŸÄ±nda 'alÄ±cÄ±'nÄ±n temel sorumluluklarÄ± nelerdir?","gt":"TODO"},
  {"id":"N09","type":"Factual","q":"TaÅŸÄ±ma evrakÄ±nda bulunmasÄ± gereken asgari bilgiler nelerdir?","gt":"TODO"},
  {"id":"N10","type":"Factual","q":"GÃ¶nderen tarafÄ±ndan hazÄ±rlanmasÄ± gereken taÅŸÄ±ma evrakÄ± hangi aÅŸamada taÅŸÄ±yana teslim edilir?","gt":"TODO"},

  {"id":"N11","type":"Factual","q":"SRC5 belgesi hangi faaliyet/taÅŸÄ±ma tÃ¼rleri iÃ§in gereklidir?","gt":"TODO"},
  {"id":"N12","type":"Factual","q":"SRC5 belgesi olmayan sÃ¼rÃ¼cÃ¼yÃ¼ Ã§alÄ±ÅŸtÄ±rmanÄ±n yaptÄ±rÄ±mÄ±/cezasÄ± nedir?","gt":"TODO"},
  {"id":"N13","type":"Factual","q":"Tehlikeli madde taÅŸÄ±macÄ±lÄ±ÄŸÄ±nda sÃ¼rÃ¼cÃ¼nÃ¼n yanÄ±nda bulunmasÄ± gereken belgeler nelerdir?","gt":"TODO"},
  {"id":"N14","type":"Factual","q":"TaÅŸÄ±ma biriminde (araÃ§ta) hangi ikaz/iÅŸaretleme kurallarÄ± uygulanÄ±r?","gt":"TODO"},
  {"id":"N15","type":"Factual","q":"Kapsam dÄ±ÅŸÄ± sayÄ±lan kurum/araÃ§ Ã¶rnekleri nelerdir?","gt":"TODO"},
  {"id":"N16","type":"Factual","q":"TMFB Ã¼cretinin gÃ¼ncel tutarÄ± nedir?","gt":"TODO"},
  {"id":"N17","type":"Factual","q":"TaÅŸÄ±ma evrakÄ±nÄ±n eksik hazÄ±rlanmasÄ±nÄ±n cezasÄ± nedir?","gt":"TODO"},
  {"id":"N18","type":"Factual","q":"AlÄ±cÄ± olarak TMFB alma zorunluluÄŸu hangi eÅŸik miktardan sonra baÅŸlar?","gt":"TODO"},
  {"id":"N19","type":"Factual","q":"TMGD hizmeti alÄ±nmasÄ± gereken iÅŸletmeler TMGDâ€™yi iÃ§ kaynak mÄ± dÄ±ÅŸ hizmet mi olarak saÄŸlayabilir?","gt":"TODO"},
  {"id":"N20","type":"Factual","q":"EÄŸitim alan personel iÃ§in belgelendirme/sertifika geÃ§erliliÄŸi nasÄ±l tanÄ±mlanÄ±r?","gt":"TODO"},

  {"id":"N21","type":"Reasoning","q":"GÃ¶nderen taÅŸÄ±ma evrakÄ±nda kritik bir bilgiyi (Ã¶r. UN numarasÄ±) yanlÄ±ÅŸ yazarsa sorumluluk kimdedir ve hangi sonuÃ§ doÄŸar?","gt":"TODO"},
  {"id":"N22","type":"Reasoning","q":"TaÅŸÄ±yan, gÃ¶nderenden evrakÄ± eksik aldÄ±ÄŸÄ± halde yola Ã§Ä±karsa hangi taraf(lar) hangi sorumluluklarÄ± ihlal eder?","gt":"TODO"},
  {"id":"N23","type":"Reasoning","q":"TMFB sahibi vefat ederse belgenin mirasÃ§Ä±ya geÃ§iÅŸ sÃ¼reci nasÄ±l iÅŸler?","gt":"TODO"},
  {"id":"N24","type":"Reasoning","q":"Bir iÅŸletme hem gÃ¶nderen hem alÄ±cÄ± rolÃ¼ndeyse TMFB/TMGD yÃ¼kÃ¼mlÃ¼lÃ¼klerini nasÄ±l deÄŸerlendirir?","gt":"TODO"},
  {"id":"N25","type":"Reasoning","q":"Kapsam dÄ±ÅŸÄ± bir araÃ§/kurumla taÅŸÄ±nan tehlikeli madde iÃ§in hangi hÃ¼kÃ¼mlerin uygulanmadÄ±ÄŸÄ±/uygulandÄ±ÄŸÄ± nasÄ±l ayrÄ±lÄ±r?","gt":"TODO"},
  {"id":"N26","type":"Reasoning","q":"EÅŸik miktarÄ±n altÄ±nda kalan bir iÅŸletme hangi ÅŸartlarda yine de TMGD/TMFB yÃ¼kÃ¼mlÃ¼lÃ¼ÄŸÃ¼ne girebilir?","gt":"TODO"},
  {"id":"N27","type":"Reasoning","q":"EÄŸitim yenileme sÃ¼resini kaÃ§Ä±ran bir iÅŸletme hangi adÄ±mlarla tekrar uyuma dÃ¶ner?","gt":"TODO"},
  {"id":"N28","type":"Reasoning","q":"SRC5â€™i olmayan sÃ¼rÃ¼cÃ¼yle taÅŸÄ±maya teÅŸebbÃ¼s edilirse taÅŸÄ±ma zincirinde hangi kontrol noktalarÄ± bunu yakalamalÄ±dÄ±r?","gt":"TODO"},
  {"id":"N29","type":"Reasoning","q":"TaÅŸÄ±ma evrakÄ± eksikse, taÅŸÄ±ma baÅŸlamadan Ã¶nce en doÄŸru dÃ¼zeltme akÄ±ÅŸÄ± (kim kime neyi iletir) nasÄ±l olmalÄ±dÄ±r?","gt":"TODO"},
  {"id":"N30","type":"Reasoning","q":"Bir denetimde hem evrak eksikliÄŸi hem de personel eÄŸitim sÃ¼resi geÃ§miÅŸse ihlaller nasÄ±l sÄ±nÄ±flanÄ±r ve Ã¶nceliklendirilir?","gt":"TODO"},

  {"id":"E01","type":"Edge/Refusal","q":"IATAâ€™ya gÃ¶re havayolu tehlikeli madde paketleme kurallarÄ±nÄ± madde madde yazar mÄ±sÄ±n?","gt":"Kapsam dÄ±ÅŸÄ± (Refusal)."},
  {"id":"E02","type":"Edge/Refusal","q":"TNTâ€™nin kimyasal formÃ¼lÃ¼ nedir?","gt":"Bilgi bulunamadÄ±."},
  {"id":"E03","type":"Edge/Refusal","q":"UN 1203 (benzin) iÃ§in ADR sÄ±nÄ±fÄ±, paketleme grubu ve tÃ¼nel kÄ±sÄ±tlama kodunu sÃ¶yle.","gt":"Bilgi bulunamadÄ±."},
  {"id":"E04","type":"Edge/Refusal","q":"2026 yÄ±lÄ± iÃ§in SRC5 sÄ±nav Ã¼creti kaÃ§ TL?","gt":"Bilgi bulunamadÄ±."},
  {"id":"E05","type":"Edge/Refusal","q":"Googleâ€™da yayÄ±mlanan en gÃ¼ncel TMGD yÃ¶netmelik deÄŸiÅŸikliklerini Ã¶zetle.","gt":"Kapsam dÄ±ÅŸÄ± (Refusal)."},
  {"id":"E06","type":"Edge/Refusal","q":"ABD DOT dÃ¼zenlemelerine gÃ¶re tehlikeli madde etiketleme kurallarÄ± nelerdir?","gt":"Bilgi bulunamadÄ±."},
  {"id":"E07","type":"Edge/Refusal","q":"Bir tehlikeli maddenin MSDS belgesinden UN numarasÄ±nÄ± otomatik Ã§Ä±karan regex yaz.","gt":"Kapsam dÄ±ÅŸÄ± (Refusal)."},
  {"id":"E08","type":"Edge/Refusal","q":"TMFB Ã¼cretini bu yÄ±lki yeniden deÄŸerleme oranÄ±na gÃ¶re gÃ¼ncelleyip yeni Ã¼creti hesapla (oranÄ± ben sÃ¶ylemiyorum).","gt":"Bilgi bulunamadÄ±."},
  {"id":"E09","type":"Edge/Refusal","q":"TMGD olmak iÃ§in gerekli Ã¼niversite bÃ¶lÃ¼mleri ve gÃ¼ncel sÄ±nav takvimi nedir?","gt":"Bilgi bulunamadÄ±."},
  {"id":"E10","type":"Edge/Refusal","q":"TÃ¼rkiyeâ€™de tehlikeli madde taÅŸÄ±macÄ±lÄ±ÄŸÄ±yla ilgili tÃ¼m kanunlarÄ± ve yÃ¶netmelikleri eksiksiz listele.","gt":"Kapsam dÄ±ÅŸÄ± (Refusal)."},
]

print(f"\n Benchmark baÅŸlÄ±yor | Model={MODEL_NAME} | Sleep={SLEEP_SEC}s | Soru={len(benchmark_data)}")
results = []

for item in tqdm(benchmark_data):
    q = item["q"]
    gt = item["gt"]
    qtype = item["type"]

    t0 = time.time()
    ans = answer_with_rag(q)
    elapsed = round(time.time() - t0, 2)

    sc, reason = score_answer(qtype, ans, gt)

    results.append({
        "ID": item["id"],
        "Type": qtype,
        "Question": q,
        "Model_Answer": ans,
        "Ground_Truth": gt,
        "Score": sc,
        "Reason": reason,
        "Time": elapsed
    })

    time.sleep(SLEEP_SEC)

df = pd.DataFrame(results)
print("\n" + "="*70)
print(f" Ortalama Skor: {df['Score'].mean():.1f}/100")
print("="*70)

out_path = f"{LOG_DIR}/benchmark_rate_limit_safe.xlsx"
df.to_excel(out_path, index=False)
print(" Excel kaydedildi:", out_path)

df

def ask(q: str):
    return answer_with_rag(q)

print("\n HazÄ±r. Ã–rnek: print(ask('TMFB Ã¼creti nedir?'))")
```
